```python
import pandas as pd

# Load the dataset
dataframe = pd.read_csv("Zomato data .csv")
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
    
```

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")
```

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
result.reset_index(inplace=True)  # To use 'listed_in(type)' as a column for plotting
plt.plot(result['listed_in(type)'], result['votes'], c="green", marker="o")  # Line plot
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)

```

```python
plt.hist(dataframe['rate'],bins =5)
plt.title("ratings distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the boxplot
plt.figure(figsize=(6, 6))
sns.boxplot(x='online_order', y='rate', data=dataframe)
```

```python
Conclusion - offline order received lower rating in comparison to offline order.
```

```python
dataframe.head()
```

```python
# Create the pivot table
pivot_table = dataframe.pivot_table( index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
# Plot the heatmap
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d')  # Use 'annot=True' to show values, 'fmt="d"' for integer formatting
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

```python
Conclusion- Dining restaurants primarily accepts offline orders ,whereas cafes primarily receive online orders .This suggest  that clients prefer 
orders in person at retaurants,but prefer online ordering at cafes.
```
